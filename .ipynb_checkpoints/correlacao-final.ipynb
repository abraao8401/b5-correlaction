```python
import numpy as np
import os
import whisper
import cv2
import pickle
from PIL import Image
from transformers import AutoTokenizer, AutoModel
from transformers import CLIPProcessor, CLIPModel
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import torch
from imblearn.over_sampling import SMOTE
import torch.nn as nn
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import normalize
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, roc_auc_score, roc_curve
from sklearn.metrics.pairwise import cosine_similarity
from scipy.stats import pearsonr, spearmanr
from scipy.spatial.distance import cosine
import matplotlib.pyplot as plt
from xgboost import XGBClassifier
```

```python
# Configurações globais
DEVICE = "cuda" if torch.cuda.is_available() else "cpu"
```

```python
# Modelos
WHISPER_MODEL = whisper.load_model("base").to(DEVICE)  # Modelo Whisper para transcrição
BERT_TOKENIZER = AutoTokenizer.from_pretrained("bert-base-uncased")
BERT_MODEL = AutoModel.from_pretrained("bert-base-uncased").to(DEVICE)

# Carregando CLIP corretamente
CLIP_MODEL = CLIPModel.from_pretrained("openai/clip-vit-base-patch32").to(DEVICE)
CLIP_PROCESSOR = CLIPProcessor.from_pretrained("openai/clip-vit-base-patch32")
```

## Pré-processamento dos Vídeos

```python
def extract_frames(video_path, frame_rate=1):
    """Extrai frames de um vídeo a uma taxa específica"""
    cap = cv2.VideoCapture(video_path)
    frames = []
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(fps / frame_rate)
    frame_count = 0

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        if frame_count % frame_interval == 0:
            frames.append(frame)
        frame_count += 1
    cap.release()
    return frames

def transcribe_audio(video_path):
    """Transcreve o áudio do vídeo usando Whisper"""
    try:
        result = WHISPER_MODEL.transcribe(video_path)
        return result["text"]
    except Exception as e:
        print(f"Erro ao transcrever {video_path}: {e}")
        return ""

# Caminhos dos vídeos
input_dirs = ["/kaggle/input/dmapf-correlacao/videos"]

# Diretório de saída
output_dir = "/kaggle/working/"
os.makedirs(output_dir, exist_ok=True)
frames_dir = os.path.join(output_dir, "frames")
transcriptions_dir = os.path.join(output_dir, "transcriptions")
os.makedirs(frames_dir, exist_ok=True)
os.makedirs(transcriptions_dir, exist_ok=True)

# Dicionário para armazenar os resultados
video_data = {}

for input_dir in input_dirs:
    for video_file in os.listdir(input_dir):
        video_path = os.path.join(input_dir, video_file)
        print(f"Processando: {video_path}")

        # Extração de frames
        frames = extract_frames(video_path)
        
        # Transcrição de áudio
        transcript = transcribe_audio(video_path)

        # Salvar frames individualmente
        frames_path = os.path.join(frames_dir, f"{video_file}.pkl")
        with open(frames_path, "wb") as f:
            pickle.dump(frames, f)

        # Salvar transcrição
        transcript_path = os.path.join(transcriptions_dir, f"{video_file}.txt")
        with open(transcript_path, "w") as f:
            f.write(transcript)

        # Armazena os dados processados
        video_data[video_file] = {
            "frames": frames,
            "transcription": transcript
        }

print("Pré-processamento concluído!")
```

## Extração de Embeddings Visuais e Textuais

```python
# Diretórios
frames_dir = "/kaggle/working/frames/"
transcriptions_dir = "/kaggle/working/transcriptions/"
embeddings_dir = "/kaggle/working/embeddings/"
os.makedirs(embeddings_dir, exist_ok=True)

def load_frames(video_id):
    """Carrega os frames salvos como arquivos pickle."""
    frame_path = os.path.join(frames_dir, f"{video_id}.pkl")
    if not os.path.exists(frame_path):
        return None
    with open(frame_path, "rb") as f:
        frames = pickle.load(f)
    return frames

def extract_clip_embeddings(frames, clip_model, clip_processor, device):
    """Extrai embeddings visuais dos frames usando CLIP."""
    clip_model.eval()
    frame_embeddings = []
    
    with torch.no_grad():
        for frame in frames:
            image = Image.fromarray(frame)
            inputs = clip_processor(images=image, return_tensors="pt").to(device)
            embedding = clip_model.get_image_features(**inputs).squeeze().cpu().numpy()
            frame_embeddings.append(embedding)
    
    # Normalização L2 dos embeddings visuais
    frame_embeddings = np.array(frame_embeddings)
    frame_embeddings = normalize(frame_embeddings, norm='l2')
    
    return frame_embeddings

def load_transcription(video_id):
    """Carrega a transcrição do vídeo."""
    transcription_path = os.path.join(transcriptions_dir, f"{video_id}.txt")
    if not os.path.exists(transcription_path):
        return None
    with open(transcription_path, "r", encoding="utf-8") as f:
        transcription = f.read().strip()
    return transcription

def extract_text_embeddings(text, tokenizer, model, device):
    """Extrai embeddings textuais usando BERT."""
    model.eval()
    with torch.no_grad():
        inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512).to(device)
        outputs = model(**inputs)
        embedding = outputs.last_hidden_state.mean(dim=1).squeeze().cpu().numpy()
    
    # Normalização L2 do embedding textual
    embedding = normalize(embedding.reshape(1, -1), norm='l2').flatten()
    
    return embedding

def process_all_videos():
    """Itera por todos os vídeos e extrai embeddings."""
    video_ids = [f.split(".pkl")[0] for f in os.listdir(frames_dir) if f.endswith(".pkl")]
    
    for video_id in video_ids:
        print(f"Processando vídeo: {video_id}")
        
        frames = load_frames(video_id)
        text = load_transcription(video_id)
        
        if frames:
            visual_embeddings = extract_clip_embeddings(frames, CLIP_MODEL, CLIP_PROCESSOR, DEVICE)
            # Salvar embeddings visuais
            with open(os.path.join(embeddings_dir, f"{video_id}_visual.pkl"), "wb") as f:
                pickle.dump(visual_embeddings, f)
            print(f"Embeddings visuais extraídos para {video_id} ({len(visual_embeddings)} frames).")
        
        if text:
            text_embedding = extract_text_embeddings(text, BERT_TOKENIZER, BERT_MODEL, DEVICE)
            # Salvar embeddings textuais
            with open(os.path.join(embeddings_dir, f"{video_id}_text.pkl"), "wb") as f:
                pickle.dump(text_embedding, f)
            print(f"Embedding textual extraído para {video_id} com shape {text_embedding.shape}.")
```

```python
process_all_videos()
```

## Balanceamento e treinamento do Dataset com o framework SMOTE

```python
# Caminho para o arquivo CSV
csv_path = "/kaggle/working/videos_labels.csv"

# Carregar os dados do CSV
video_data_df = pd.read_csv(csv_path)

def load_features():
    """Carrega as características (embeddings) dos vídeos."""
    features = []

    for video_id in video_data_df['video']:
        visual_embedding_path = os.path.join(embeddings_dir, f"{video_id}_visual.pkl")
        text_embedding_path = os.path.join(embeddings_dir, f"{video_id}_text.pkl")
        
        if os.path.exists(visual_embedding_path) and os.path.exists(text_embedding_path):
            with open(visual_embedding_path, "rb") as f:
                visual_embedding = pickle.load(f)
            
            with open(text_embedding_path, "rb") as f:
                text_embedding = pickle.load(f)
            
            combined_embedding = np.concatenate((visual_embedding.mean(axis=0), text_embedding))  # Combina as características
            features.append(combined_embedding)
        else:
            print(f"Embeddings não encontrados para: {video_id}")

    return np.array(features)

# Carregar características e rótulos
X = load_features()
y = video_data_df['label'].values  # Rótulos do CSV

# Aplicar SMOTE antes da divisão
smote = SMOTE(sampling_strategy='auto', random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

print(f"Distribuição original: {np.bincount(y)}")
print(f"Distribuição após SMOTE: {np.bincount(y_resampled)}")

# Dividir em conjuntos de treino e teste
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Treinar o modelo XGBoost
model = XGBClassifier(
    objective='binary:logistic',
    n_estimators=200,
    max_depth=5,
    learning_rate=0.01,
    random_state=42
)
model.fit(X_train, y_train)

# Avaliação
y_pred = model.predict(X_test)
print("\n=== Relatório de Classificação ===")
print(classification_report(y_test, y_pred))
print("AUC-ROC:", roc_auc_score(y_test, y_pred))

# Ajuste de Threshold (opcional)
y_probs = model.predict_proba(X_test)[:, 1]
optimal_threshold = 0.4  # Ajuste conforme necessário (ex.: via curva ROC)
y_pred_adj = (y_probs > optimal_threshold).astype(int)
print("\n=== Relatório com Threshold Ajustado (0.4) ===")
print(classification_report(y_test, y_pred_adj))
print("AUC-ROC (ajustado):", roc_auc_score(y_test, y_pred_adj))

# Salvar resultados
results_df = pd.DataFrame({
    "video_id": video_data_df["video"],
    "target": y,  # Rótulos do CSV
    "features": list(X)  # embeddings combinados
})
results_df.to_csv(os.path.join(output_dir, "results_with_features.csv"), index=False)
print("Resultados salvos em 'results_with_features.csv'!")
```

## Visualização do Balanceamento antes e após o SMOTE

```python
# Contagem de classes antes e depois do SMOTE
original_counts = np.bincount(y)
resampled_counts = np.bincount(y_resampled)

# Criar gráfico de barras
labels = ['Alta Correlação', 'Baixa Correlação']
x = np.arange(len(labels))

fig, ax = plt.subplots(figsize=(8, 5))
bar_width = 0.35

# Barras antes do SMOTE
ax.bar(x - bar_width/2, original_counts, width=bar_width, label="Original", color='blue')

# Barras depois do SMOTE
ax.bar(x + bar_width/2, resampled_counts, width=bar_width, label="SMOTE", color='orange')

# Rótulos e título
ax.set_xlabel("Classe")
ax.set_ylabel("Número de Amostras")
ax.set_title("Distribuição das Classes Antes e Depois do SMOTE")
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

# Mostrar gráfico
plt.show()
```

## Matriz de Confusão - Curva ROC - Distribuição das probabilidades preditas

```python
# 1. Matriz de Confusão
def plot_confusion_matrix(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Fake', 'Real'], yticklabels=['Fake', 'Real'])
    plt.ylabel('Real')
    plt.xlabel('Predito')
    plt.title('Matriz de Confusão')
    plt.show()

# 2. Curva ROC
def plot_roc_curve(y_true, y_probs):
    fpr, tpr, thresholds = roc_curve(y_true, y_probs)
    plt.figure(figsize=(8, 6))
    plt.plot(fpr, tpr, label='Curva ROC (AUC = {:.2f})'.format(roc_auc_score(y_true, y_probs)))
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('Taxa de Falsos Positivos')
    plt.ylabel('Taxa de Verdadeiros Positivos')
    plt.title('Curva ROC')
    plt.legend(loc='lower right')
    plt.show()

# 3. Distribuição das Probabilidades Preditas
def plot_prediction_distribution(y_probs):
    plt.figure(figsize=(8, 6))
    sns.histplot(y_probs, bins=30, kde=True)
    plt.axvline(0.5, color='red', linestyle='--', label='Threshold = 0.5')
    plt.title('Distribuição das Probabilidades Preditas')
    plt.xlabel('Probabilidade de Classe Real')
    plt.ylabel('Frequência')
    plt.legend()
    plt.show()

# Gerar os gráficos após o treinamento do modelo
plot_confusion_matrix(y_test, y_pred)
plot_roc_curve(y_test, y_probs)
plot_prediction_distribution(y_probs)
```

## Média da Correlação de Pearson e Spearman, bem como a Média da Distância do Cosseno

```python
def calculate_correlation_metrics(X_text, X_visual):
    """Calcula métricas de correlação entre embeddings textuais e visuais."""
    pearson_corrs = []
    spearman_corrs = []
    cosine_distances = []
    
    for text_emb, visual_emb in zip(X_text, X_visual):
        # Pearson
        pearson_corr, _ = pearsonr(text_emb, visual_emb)
        pearson_corrs.append(pearson_corr)

        # Spearman
        spearman_corr, _ = spearmanr(text_emb, visual_emb)
        spearman_corrs.append(spearman_corr)

        # Distância do Cosseno
        cos_dist = cosine(text_emb, visual_emb)
        cosine_distances.append(cos_dist)
    
    return pearson_corrs, spearman_corrs, cosine_distances

# Separar embeddings textuais e visuais
X_visual = X_resampled[:, :512]  # Assume que os primeiros 512 são visuais
X_text = X_text[:, :512]  # Mantém apenas as primeiras 512 dimensões

# Calcular as métricas de correlação
pearson_corrs, spearman_corrs, cosine_distances = calculate_correlation_metrics(X_text, X_visual)

# Exibir médias das métricas
print(f"Média da Correlação de Pearson: {np.mean(pearson_corrs):.4f}")
print(f"Média da Correlação de Spearman: {np.mean(spearman_corrs):.4f}")
print(f"Média da Distância do Cosseno: {np.mean(cosine_distances):.4f}")
```
